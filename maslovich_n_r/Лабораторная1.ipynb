```python
import numpy as np

%matplotlib inline
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('vis1.jpg')
```

```python
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
```

```python
thresh_b = 150
lower_bright = np.array([0,0,thresh_b])
upper_bright = np.array([180,255,255])
```

```python
thresh_s = 100
lower_satur = np.array([0,thresh_s,0])
upper_satur = np.array([180,255,255])
```

```python
mask_b = cv.inRange(hsv, lower_bright, upper_bright)
    # Bitwise-AND mask and original image
res_b = cv.bitwise_and(img,img, mask= mask_b)
```

```python
a_b = cv.cvtColor(res_b, cv.COLOR_BGR2RGB)
plt.imshow(a_b)
```

```python
mask_s = cv.inRange(hsv, lower_satur, upper_satur)
    # Bitwise-AND mask and original image
res_s = cv.bitwise_and(img,img, mask= mask_s)
```

```python
a_s = cv.cvtColor(res_s, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(a_s)
```

```python
res = cv.bitwise_and(res_b, res_s)
a = cv.cvtColor(res, cv.COLOR_BGR2RGB)
plt.imshow(a)
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
lower_satur = np.array([0,10,0])
upper_satur = np.array([180,255,255])
mask_s = cv.inRange(hsv, lower_satur, upper_satur)
# Bitwise-AND mask and original image
res_s = cv.bitwise_and(img,img, mask= mask_s)
a_s = cv.cvtColor(res_s, cv.COLOR_BGR2RGB)
```

```python
def apply_saturation_thresh(thresh_s,axer):
    lower_satur = np.array([0,thresh_s,0])
    upper_satur = np.array([180,255,255])
    mask_s = cv.inRange(hsv, lower_satur, upper_satur)
    # Bitwise-AND mask and original image
    res_s = cv.bitwise_and(img,img, mask= mask_s)
    a_s = cv.cvtColor(res_s, cv.COLOR_BGR2RGB)
    axer.set_data(a_s)
    plt.imshow(a_s)
```

```python
slider = widgets.IntSlider()
display(slider)
ax = plt.imshow(a_s)
def on_value_change(change):
    apply_saturation_thresh(change['new'],ax)
slider.observe(on_value_change, names='value')
```
