## 2D-Convolution using OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (9.0, 9.0)
```

```python
img = cv2.imread('vis2.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
plt.imshow(gray[200:400,550:800], cmap = 'gray')
#plt.axis("off")
plt.show()
```

```python
out = gray
kernel = np.array([-0.5, 0.5]).reshape(1,2)
out = cv2.filter2D(out,-1,kernel)
```

```python
plt.imshow(out[200:400,550:800], cmap = 'gray')
plt.axis("off")
plt.show()
```
